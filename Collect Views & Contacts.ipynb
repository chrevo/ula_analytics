```python
# Transform Region to city

regions_map = {'D09AD180D0B0D181D0BDD0BED0B4D0B0D180D181D0BAD0B8D0B920D0BAD180D0B0D0B9':1,
               'D0A1D0B2D0B5D180D0B4D0BBD0BED0B2D181D0BAD0B0D18F20D0BED0B1D0BBD0B0D181D182D18C':2,
               'D0A0D0BED181D182D0BED0B2D181D0BAD0B0D18F20D0BED0B1D0BBD0B0D181D182D18C':3,
               'D0A2D0B0D182D0B0D180D181D182D0B0D0BD':4}

searches['region_id'] = searches['region_id'].map(regions_map)
views['region_id'] = views['region_id'].map(regions_map)
contacts['region_id'] = contacts['region_id'].map(regions_map)

# Fillna and transform to int

for col in tqdm(searches.columns[-12:]):
    searches[col] = searches[col].fillna(0).astype(str).apply(lambda x: x.replace(',','.')).astype(float)
    searches[col] = searches[col].astype(int)
    
searches['default'] = (searches['default'] == 'main').astype(int)
```

### Need to merge and fill NA with the help of views and contacts

```python
views = views.rename(columns= {'subcategory_id':'subcat_view',
                               'category_id':'cat_view',
                               'cnt':'count_views'})
```

```python
cat_na_search = searches[(searches['category_id'].isna()) & (~searches['subcategory_id'].isna())].reset_index(drop = True)
subcat_na_search = searches[(~searches['category_id'].isna()) & (searches['subcategory_id'].isna())].reset_index(drop = True)
both_na_search = searches[(searches['category_id'].isna()) & (searches['subcategory_id'].isna())].reset_index(drop = True)

search_no_miss = searches.dropna(subset= ['category_id', 'subcategory_id']).reset_index(drop = True)
```

```python
views.head()
```

```python
cat_na_search.head()
```

```python
# 1. Add missing parent category
unique_pairs = searches.groupby(['category_id', 'subcategory_id']).size().reset_index()
del unique_pairs[0]
unique_pairs.columns = ['cat_view', 'subcategory_id']

cat_na_search = pd.merge(cat_na_search, unique_pairs, on = 'subcategory_id', how = 'left')
cat_na_search['category_id'] = cat_na_search['category_id'].fillna(cat_na_search['cat_view'])
del cat_na_search['cat_view']
```

```python
cat_na_search.head()
```

```python
# 2. Fill missing category
```

```python
views_1 = views.rename(columns ={'cat_view':'category_id'}).copy()
subcat_na_search = pd.merge(subcat_na_search, views_1.drop_duplicates(subset = ['user_id', 'search_id', 'category_id'], keep ='last'), 
         on = ['region_id', 'user_id', 'search_id', 'category_id'], how = 'left')

subcat_na_search['subcategory_id'] = subcat_na_search['subcategory_id'].fillna(subcat_na_search['subcat_view'])
subcat_na_search['subcategory_id'] = subcat_na_search['subcategory_id'].fillna('-1')

del subcat_na_search['subcat_view'], subcat_na_search['count_views'], views_1
```

```python
print(subcat_na_search.shape)
subcat_na_search.head()
```

```python
# 3. Fill where both are missed
```

```python
both_na_search = pd.merge(both_na_search, views.drop_duplicates(subset = ['user_id', 'search_id'], keep ='last'), 
                          on = ['region_id', 'user_id', 'search_id'], how = 'left')

both_na_search['subcategory_id'] = both_na_search['subcategory_id'].fillna(both_na_search['subcat_view'])
both_na_search['subcategory_id'] = both_na_search['subcategory_id'].fillna('-1')

both_na_search['category_id'] = both_na_search['category_id'].fillna(both_na_search['cat_view'])
both_na_search['category_id'] = both_na_search['category_id'].fillna('-1')

del both_na_search['subcat_view'], both_na_search['count_views'], both_na_search['cat_view']
```

```python
both_na_search.head()
```

```python
# Adding everythin together

searches = pd.concat([search_no_miss, cat_na_search, subcat_na_search, both_na_search], ignore_index = True)
```

```python
searches
```

### Add views and contacts

```python
searches = pd.merge(searches, views.rename(columns = {'cat_view':'category_id',
                                           'subcat_view':'subcategory_id'}), 
         on = ['subcategory_id', 'category_id', 'search_id', 'user_id', 'region_id'],
         how = 'left')

searches['count_views'] = searches['count_views'].fillna(0).astype(int)
```

```python
searches.head()
```

```python
contacts.head()
```

```python
df = pd.merge(searches, contacts, 
         on = ['subcategory_id', 'category_id', 'search_id', 'user_id', 'region_id'],
         how = 'left')

df['count_contacts'] = df['cnt'].fillna(0).astype(int)
del df['cnt']
```

```python
df
```

```python
# Viewed and contacted marks
df['viewed'] = (df['count_views'] > 0).astype(int)
df['contacted'] = (df['count_contacts'] > 0).astype(int)
```

```python
df.to_csv('prepared_data.csv', index = False)
```
